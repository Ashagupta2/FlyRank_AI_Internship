## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule like "flag all pages where trend_direction == down" ignores impressions volume, content age, and how these factors interact. My earlier baseline test showed a hand-written rule gets precision@50 = 0.24, while a random forest model reaches 0.74 — a 3x improvement. This gap shows the real pattern is multi-dimensional and non-linear, which a single if-statement can't capture but a learned model can.
